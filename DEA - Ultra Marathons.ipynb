```python
df[df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA']
```

```python
# combine all the filters together
```

```python
df[(df['Event distance/length'].isin(['50km', '50mi'])) & \
    (df['Year of event'] == 2020) & \
    (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & \
    (df['Year of event'] == 2020) & \
    (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.head(10)
```

```python
df2.shape
```

```python
# remove (USA) from event name
```

```python
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head(10)
```

```python
# Clean up "Athlete age category"
```

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
# remove H from athlete performace
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head()
```

```python
# drop columns
# Athlete Club, Athlete Country, Athlete year of birth, Athlete age category.
```

```python
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
df2.head()
```

```python
# clean up null values
```

```python
df2.isna().sum()
```

```python
df2[df2['athlete_age'].isna()==1]
```

```python
#drop the rows with null values.
```

```python
df2 = df2.dropna()
```

```python
df2.shape
```

```python
# check for dupes
```

```python
df2[df2.duplicated() == True]
```

```python
#reset index
```

```python
df2.reset_index(drop = True)
```

```python
#fix types
```

```python
df2.dtypes
```

```python
df2['athlete_age'] = df2['athlete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
df2.dtypes
```

```python
df2.head()
```

```python
df2 = df2.rename(columns = {'Year of event':'year',
                            'Event dates':'race_day',
                            'Event name':'race_name',
                            'Event distance/length':'race_length',
                            'Event number of finishers':'race_number_of_finishers',
                            'Athlete performance':'athlete_performace',
                            'Athlete gender':'athlete_gender',
                            'Athlete average speed':'athlete_average_speed',
                            'Athlete ID':'athlete_id'
})
```

```python
df2.head()
```

```python
# reorder columns
```

```python
df3 = df2[['race_day', 'race_name', 'race_length', 'race_number_of_finishers', 'athlete_id', 'athlete_gender', 'athlete_age', 'athlete_performace', 'athlete_average_speed']]
```

```python
# questions I want to find out from the data
```

```python
# difference in speed for the 50k, 50mi male to female
```

```python
df3.groupby(['race_length', 'athlete_gender'])['athlete_average_speed'].mean()
```

```python
# what age groups are the best in the 50mi race (20 + races min)
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count>19')
```

```python
# what age groups are the worst
```

```python
df3.query('race_length == "50mi"').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = True).query('count>10')
```

```python
# seasons for the data, are runners slower in summer than winter?

#spring 3-5
#summer 6-8
#fall 9-11
#winter 12-2

#split between two decimals
```

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: "Winter" if x > 11 else "Fall" if x > 8 else "Summer" if x > 5 else "Spring" if x > 2 else "Winter")
```

```python
df3.head(25)
```

```python
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

```python
# 50 mile only
```

```python
df3.query('race_length == "50mi"').groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```
